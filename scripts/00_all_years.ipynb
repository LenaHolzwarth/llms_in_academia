```python
import os
import pickle
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
path = "../data/baseline_2026-01-23/formatted/"
```

```python
dfs = []
paper_count = 0
for year in tqdm(range(2000, 2026)):
    df = pd.read_json(path + str(year) + ".json")
    paper_count += len(df)
    dfs.append(df)

big_df = pd.concat(dfs, ignore_index=True)
```

```python
print("big_df memory (approx MB):", big_df.memory_usage(deep=True).sum()/2**20)
```

```python
big_df.to_pickle(os.path.join(path, "all.pkl")) 
```
