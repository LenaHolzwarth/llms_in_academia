### Notebook to test the download and preprocessing of PubMedCentral data

Relevant PMC subset is PMC Open Access Subset
adress: https://ftp.ncbi.nlm.nih.gov/pub/pmc/oa_bulk/

there are 3 folders (each have .txt and .xml subfolders, we use xml):
- oa_comm/xml (commercial use): baseline ~ 90GB compressed (comm + noncomm extracted ~ 580 GB)
- oa_noncomm/xml (non-commercial use): baseline ~28GB
- oa_other/xml (no machine-readable license, no license or custom license)

each folder has baseline (all articles up to a certain date) and increment (daily updated of papers added after the baseline) file packages. Each package also has a file list (in .csv and .txt)
- file packages: ending .tar.gz (need to be extracted with tar -xzf file.tar.gz)
- file list: ending .filelist.csv / .filelist.txt
- baseline: 'oa_comm_xml.PMC012xxxxxx.baseline.2025-06-26.tar.gz' 
- increment: 'oa_comm_xml.incr.2025-06-27.tar.gz'

```python
from ftplib import FTP
import os
import re 
import requests
from tqdm import tqdm
```

```python
test_filelist = 'oa_comm_xml.PMC012xxxxxx.baseline.2025-06-26.filelist.csv'
test_package = 'oa_comm_xml.PMC012xxxxxx.baseline.2025-06-26.tar.gz'
```

```python
DATA_PATH = "/gpfs01/berens/data/data/pubmed_central/baseline_2026-01-23"
FILENAME = "oa_comm_xml.PMC006xxxxxx.baseline.2026-01-23.tar.gz"
```

```python
url = "https://ftp.ncbi.nlm.nih.gov/pub/pmc/oa_bulk/oa_comm/xml/" + FILENAME
r = requests.get(url, stream=True)
```

```python
path = os.path.join(DATA_PATH, FILENAME)
if not os.path.exists(path):
    total_size = int(r.headers.get("content-length", 0))

    with open(path, "wb") as f, tqdm(
        total=total_size,
        unit="B",
        unit_scale=True,
        unit_divisor=1024,
    ) as bar:
        for chunk in r.iter_content(chunk_size=8192):
            if chunk:
                f.write(chunk)
                bar.update(len(chunk))
else:
    print(f"file {FILENAME} already downloaded")
```
