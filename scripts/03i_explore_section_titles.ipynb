Notebook to investigate for different article-types, which are the common section names

article types investigated:
- 'research-article'
- 'review-article'
- 'case-report'

Goal: get a dictionary for each article type that standardizes commonly used section names

baseline sections used to create dict:
- 012 comm
- 008 noncomm
- 004 comm
- 000 comm

```python
import pandas as pd
import re
from collections import Counter

%load_ext autoreload
%autoreload 2
import src.data_utils as tv
```

```python
baseline_df = pd.read_json("../data/baseline_2025-06-26/PMC008xxxxxx_noncomm.json")
paper_count_total = len(baseline_df)
paper_count_total
```

```python
# article-type investigated
article_type = "research-article"
```

```python
# crop df to only include article-type
df = baseline_df[baseline_df["article-type"] == article_type]
paper_count_cropped = len(df)
paper_count_cropped
```

```python
# proportion of article type in baseline
paper_count_cropped / paper_count_total
```

```python
# get all section titles
titles = [x for y in df["section_titles"].dropna() for x in y if not x == None]
titles = list(map(str.casefold, titles))

# clean titles
for i, t in enumerate(titles):
    for pat, repl in tv.title_clean.items():
        t = re.sub(pat, repl, t)
    titles[i] = t.strip(":").strip()
```

```python
Counter(titles)
```

```python
titles_aggregated = []
for t in titles:
    alt_t = t
    for key, synonyms in tv.title_versions.items():
        if t in synonyms:
            alt_t = key
    titles_aggregated.append(alt_t)
```

```python
Counter(titles_aggregated)
```
